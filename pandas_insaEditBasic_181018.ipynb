```python
# 판다스 임포트 #pandas
import pandas as pd
from pandas import Series, DataFrame
import re
```

## csv파일 정리하기

```python
#기수, 직위, 성명, 성별, 전화번호, 비고 가져오기
df.columns = ['1', '학교명','이름','수정사항','2','전보유형','3']

# 불필요한 열 삭제하기
df = df.drop(['1', '2','3'], axis = 1)
```

```python
df.head()
```

```python
# 연락처 빈칸인 행 삭제 #pandas #느낀점
dfSel = df.dropna(thresh=2) #이름해이 빈칸이면 삭제
dfSel.head()
```

#### '학교명' 값이 없는 행들만 저장하기

```python
dfSel = dfSel[dfSel['학교명'] != '학교명']
dfSel
```

#### 정렬하기

```python
dfSel = dfSel.sort_values(by = ['전보유형','학교명','이름'])
dfSel.head()
```

#### 인덱싱 다시 하기

```python
dfSel.reset_index(drop=True, inplace=True)
dfSel.head()
```

#### 중복값 삭제

```python
dfSel = dfSel.drop_duplicates('수정사항')
```

#### 파일저장

```python
dfSel.to_excel('pandas_insaEditFinal_181018.xlsx')
```
