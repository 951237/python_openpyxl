## 데이터 프레임 만들기(전체)

```python
# 엑셀파일 불러서 데이터 프레임 만들기
df = pd.read_excel(inputFile)

# 2행부터 마지막 행까지 선택하기 loc[선택행, 선택열]
df = df.loc[2:,:]

# 인덱스 번호 재설정하기 
df.reset_index(drop=True, inplace=True)

# 칼럼 열 이름 재설정
df.columns = ['그룹','학년','반','학교이름','전화','메모']
df.head()
```
