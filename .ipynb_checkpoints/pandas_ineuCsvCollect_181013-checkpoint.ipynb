```python
# 판다스 임포트 #pandas
import pandas as pd
from pandas import Series, DataFrame
import re
```

```python
# path 내의 절대 경로 가져오기 함수 #pandas #느낀점 참고만 하기 
#glob 기능이 더 좋음.
import os

def allfiles(path):
    res = []

    for root, dirs, files in os.walk(path):
        rootpath = os.path.join(os.path.abspath(path), root)

        for file in files:
            filepath = os.path.join(rootpath, file)
            res.append(filepath)

    return res
```

```python
#하위폴더 파일 부르기
inputFile = './ineu_csv_181003/pandas_ineuMac.csv'
```

## csv파일 정리하기

```python
df = pd.read_csv(inputFile )

#기수, 직위, 성명, 성별, 전화번호, 비고 가져오기
df.columns = ['num', 'gradYear','rank','name','phone','sex','ifMoney','ifAttend','memo']

# 불필요한 열 삭제하기
df = df.drop(['ifMoney', 'ifAttend'], axis = 1)

# 연락처 빈칸인 행 삭제 #pandas #느낀점
dfSel = df.dropna(thresh=6)

#소속 추가하기
dfSel['school'] = ''

dfSel
```

```python
dfEdit = dfSel
```

```python
# 조건을 검색하여 특정행 삭제하기 
# 조건에 맞는 행을 검색 → 인덱스를 따서 데이터 프레임에서 삭제
dfDel = dfEdit.loc[(dfEdit['num'].str.contains('연번'))]
dfEdit.drop(dfDel.index, inplace=True)
```

```python
# 직위로 정렬후 인덱싱 다시 하기
dfEdit=dfEdit.sort_values(by = ['rank', 'name'])
dfEdit.reset_index(drop=True, inplace=True)
dfEdit.head()
```

```python
# num컬럼 행 삭제하기 #pandas
dfEdit.drop('num',axis=1, inplace=True)
```

## Groupby 기능 활용하기

```python
# 직위로 묶어서 변수에 넣기
groupbyRank = dfEdit.groupby('rank')
```

```python
# 그룹을 개괄적으로 확인하기
groupbyRank.groups
```

```python
# 반복문을 이용하여 그룹을 출력하기 #pandas #느낀점 
for name, group in groupbyRank:
    print(name + " : " + str(len(group)))
    print(group)
    print()
```

```python
# 행에서 특정 조건에 맞는 문자열 검색하기 #느낀점 #pandas
dfTeacher = dfEdit.loc[(dfEdit['rank'].str.contains('교사'))]
dfTeacher.head()
```

```python
# 행에서 특정 조건에 맞는 문자열 검색하기 #느낀점 #pandas
dfSubPrinciple = dfEdit.loc[(dfEdit['rank'].str.contains('교감'))]
dfSubPrinciple.head()
```

```python
# 행에서 특정 조건에 맞는 문자열 검색하기 #느낀점 #pandas
dfPrinciple = dfEdit.loc[(dfEdit['rank'].str.contains('교장'))]
dfPrinciple.head()
```
