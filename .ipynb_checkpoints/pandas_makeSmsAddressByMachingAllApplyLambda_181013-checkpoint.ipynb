```python
# 판다스 임포트 #pandas
import pandas as pd
from pandas import Series, DataFrame
import pdb
```

```python
# 불러올 파일 지정하기 #pandas
inputFile1 = 'pandas_googleAddress_181012.csv'
```

```python
# 불러올 파일 지정하기 #pandas
inputFile2 = 'pandas_matchSms.csv'
```

```python
# CSV파일을 df(데이터 프레임)로 부르기 #pandas
dfFrom = pd.read_csv(inputFile1, encoding='utf-8', index_col=0)
```

```python
# 열 이름 바꾸기 #느낀점 #pandas
dfFrom.columns = ['name', 'phone', 'otherPhone', 'note', 'email', 'otherEmail']
```

```python
# CSV파일을 df(데이터 프레임)로 부르기 #pandas
dfTo = pd.read_csv(inputFile2, encoding='utf-8', index_col=0)

dfTo.columns = ['name','phone']

# 데이터 프레임 초기화하여 반영하기
dfTo.reset_index(drop=True, inplace=True)
```

```python
# 왼쪽 프레임을 기준으로 값을 채워 넣음. #느낀점
dfTo = dfTo.merge(dfFrom[['name','phone']], on = ['name'], how = 'left')
```
